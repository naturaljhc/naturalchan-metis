```python
def clean(df, NA_dict, all_dicts):
    df.replace(NA_dict, inplace = True)
    for c in df.columns:
        if c in all_dicts.keys():
            df[c].replace(all_dicts[c], inplace = True)
    drop_rows = df[(df["ANXIOUS"] == "N/A") | (df["WORRY"] == "N/A") | (df["INTEREST"] == "N/A") | (df["DOWN"] == "N/A")].index
    df.drop(drop_rows, axis = 0, inplace = True)
    output_df = df.sample(sample_num).reset_index(drop = True)
    return output_df
```

```python
df = pd.read_csv('p3.2/pulse2021_puf_34.csv')
sample_num = df.shape[0]//100
```

```python
for k in all_dicts.keys():
    if k not in df.columns:
        print(k)
```

```python
smaller_df = df.loc[:, keep_cols]
output_df = clean(smaller_df, NA_dict, all_dicts)
```
