```python
import pandas as pd
import os
import shutil
from IPython.display import Image, clear_output, display
```

```python
img_folder = 'test/'
files = os.listdir(img_folder)
files.remove('no_glasses')
files.remove('eyeglasses')
files.remove('sunglasses')
i = 1
for img in files:
    print(f"Image {i} of {len(files)}")
    print(img)
    display(Image(filename = img_folder + img, height= 400, width= 400))
    label = input("Input label:")
    if int(label) == 0:
        shutil.move(img_folder + img, img_folder + 'no_glasses/' + img)
    if int(label) == 1:
        shutil.move(img_folder + img, img_folder + 'eyeglasses/' + img)
    if int(label) == 2:
        shutil.move(img_folder + img, img_folder + 'sunglasses/' + img)
    if int(label) == 3:
        os.remove(img_folder + img)
    clear_output(wait = True)
    i += 1
```
